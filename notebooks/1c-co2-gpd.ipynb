```python
def show_labels(labels):
    size = (16, 16)
    ax = work_pd[labels[0]].plot(legend=labels[0], figsize=size, linewidth=4)
    for l in labels[1:]:
        work_pd[l].plot(legend=l, figsize=size, ax=ax, linewidth=4)
```
