```
from summarizer import Summarizer
from pprint import pprint

model_distill = Summarizer(model='distilbert-base-uncased')
```

```
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
root_dir = "/content/gdrive/My Drive/"
base_dir = root_dir + 'workshop/'
```

```
sapiens_text_file = '/content/gdrive/My Drive/sapiens.txt'

with open(sapiens_text_file) as fp:
  sapiens = fp.read()
```

```
%%time 

sapiens_short = model_distill(sapiens, ratio=0.01)
```

```
sapiens_short_file = '/content/gdrive/My Drive/sapiens-short.txt'

with open(sapiens_short_file, 'w') as fp:
  fp.write(sapiens_short)
```
