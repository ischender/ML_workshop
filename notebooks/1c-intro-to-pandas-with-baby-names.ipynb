```python
years = range(1880,2019)
pieces = []
columns = ['name','sex','births']
```

```python
for year in years:
    path = f'yob{year}.txt'
    frame = pd.read_csv(path, names = columns)
    frame['year'] = year
    pieces.append(frame)
```

```python
# we can do it better
```

```python
from glob import glob
import re

dfs = []

for file in glob('yob*.txt'):
    frame = pd.read_csv(file, names = columns)
    year = re.findall('[0-9]{4}', file)[0]
    frame['year'] = year
    dfs.append(frame)
```

```python
len(dfs), len(pieces)
```

```python
for df in dfs[:5]:
    print(df.head(3))
```

```python
names = pd.concat(dfs, ignore_index=True)
names
```

```python
total_births = names.pivot_table('births', index='year',columns = 'sex', aggfunc=sum)
total_births.tail()
```

```python
%matplotlib inline
total_births.plot(title = 'Total births by sex and year', figsize=(12,5))
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group
```

```python
names = names.groupby(['year','sex']).apply(add_prop)
```

```python
names
```

```python
def get_top(group, num=1000):
    group.sort_values(by='births',ascending=False)[:num]
    return group.sort_values(by='births',ascending=False)[:num]

get_top(names).head()
```

```python
from glob import glob
import re

dfs = []
```

```python
for file in glob('yob*.txt'):
    frame = pd.read_csv(file, names = columns)
    year = re.findall('[0-9]{4}', file)[0]
    frame[year] = frame['births']
    frame.drop(['births'], inplace=True, axis=1)
    # with_years[year] = frame[]
    dfs.append(frame)
```

```python
with_years = dfs[0]
for year_df in dfs[1:]:
  with_years = with_years.combine_first(year_df)
with_years
```
